```python
dfR= pd.read_csv('soil_moisture_groundwater_2017.csv',parse_dates=['Date time (standard time)'],
                 sep=',',skiprows=0,low_memory=False, na_values=-9999, date_parser=dateparseM)
dfR.index = dfR['Date time (standard time)']
dfR = dfR.iloc[:,1:]
dfR#.head()
```

```python
dfR.columns = ['riparian soil moisture (%)','hilllsope toe soil moisture (%)','middle hillslope soil moisture (%)','upper hillsope soil moisture (%)','GW 1','GW 4']
```

```python
dfR[['riparian soil moisture (%)','hilllsope toe soil moisture (%)','middle hillslope soil moisture (%)','upper hillsope soil moisture (%)']].plot()
```

```python
dfR.head()
```

```python
dfR['soil moisture'] = dfR[['riparian soil moisture (%)','hilllsope toe soil moisture (%)','middle hillslope soil moisture (%)','upper hillsope soil moisture (%)']].mean(axis=1)
```

```python
dfR['GW'] = dfR[['GW 1','GW 4']].mean(axis=1)
```

```python
dfR
```

```python
dfR[['soil moisture']].plot()
```

```python
dfR[['GW']].plot()
```

```python
dfR['Storage'] = dfR['soil moisture']*-10*dfR['GW']
```

```python
dfR['Storage'].plot()
```
