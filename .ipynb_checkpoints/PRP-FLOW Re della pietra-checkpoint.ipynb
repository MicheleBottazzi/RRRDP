```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dfFlow= pd.read_csv('data/RioRessi/1/RioRessi_FLOW.csv',sep=',',skiprows=2,low_memory=False)
dfPrp= pd.read_csv('data/RioRessi/1/RioRessi_PRP.csv',sep=',',skiprows=6,low_memory=False)
df2 = dfFlow
```

```python
dfFlow.columns = ['','Datetime','flow','n']
dfPrp.columns = ['','Datetime','prp']
#df[df ==-9999]=np.nan
```

```python
dfFlow['Datetime'] = pd.to_datetime(dfFlow['Datetime'], dayfirst=True)
dfFlow['Datetime'] = dfFlow['Datetime'].dt.strftime('%Y-%m-%d %H:%M')

dfPrp['Datetime'] = pd.to_datetime(dfPrp['Datetime'], dayfirst=True)
dfPrp['Datetime'] = dfPrp['Datetime'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
dfPrp.index = dfPrp['Datetime']
dfPrp.index = pd.DatetimeIndex(dfPrp.index)
dfFlow.index = dfFlow['Datetime']
dfFlow.index = pd.DatetimeIndex(dfFlow.index)
```

```python
maskYear = (dfPrp['Datetime'] >= '2012-08-20 19:00') & (dfPrp['Datetime'] <= '2016-11-23 11:00')
```

```python
dfPrp2 = dfPrp.iloc[:,1:].loc[maskYear]
dfFlow2 = dfFlow.iloc[:,1:].loc[maskYear]
```

```python
dfFlow2
```

```python
df = pd.concat([dfFlow2,dfPrp2.iloc[:,1:]],axis=1)
df.head()
```

```python
df[df == -9999]=np.nan
```
